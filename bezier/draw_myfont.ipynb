```python
from __future__ import print_function, division
import math
import copy
import turtle

# control polygon
cpList = [
    [(0,0), (100,200), (300,250), (400,0)],                         #0
    [(0,0), (100,-200), (300,250), (400,0)],                        #1
    [(0,0), (300,250), (400,-100), (200,0)],                        #2
    [(0,0), (400,300), (-200,350), (200,0)],                        #3
    [(0,0), (100,200), (300,250)],                                  #4
    [(0,0), (100,200), (300,250), (400,0)],                         #5
    [(0,0), (100,200), (300,250), (400,0), (300, -50)],             #6
    [(0,0), (100,200), (300,250), (400,0), (300, -50), (150, 50)]   #7
]

for j in range(len(cpList)): 
    for i in range(len(cpList[j])): 
        cpList[j][i] = (cpList[j][i][0] - 200, cpList[j][i][1] - 100)

# draw line segments
def draw_segments(ttl, points, color):
    ttl.penup()
    ttl.goto(points[0])
    ttl.color(color)
    ttl.pendown()
    for (x, y) in points:
        ttl.goto(x, y)

def conbination(n, k):
    return math.factorial(n) / (math.factorial(k) * math.factorial(n - k))
    
# calculate sequence of points on Bezier curve
def bezier(cp, ndot):
    # cp: controll polygon
    # ndot: number of dots
    bp = [] # dots on Bezier curve
    # calculate the coordinates of dots
    for i in range(ndot + 1):
        t = i / ndot
        l = len(cp)
        bp.append((
            sum(cp[i][0] * (1-t)**(l-i-1) * t**i * conbination(l-1, i) for i in range(l)),
            sum(cp[i][1] * (1-t)**(l-i-1) * t**i * conbination(l-1, i) for i in range(l))
        ))
    return(bp)
    
# setup turtle
gamera = turtle.Turtle()
gamera.reset()
gamera.speed(0)
# draw control polygon
for i in range(len(cpList)): draw_segments(gamera, cpList[i], "black")
# calculate dots on Bezier curve
bp = []
for i in range(len(cpList)): bp.append(bezier(cpList[i], 20))
# draw Bezier curve
for i in range(len(bp)): draw_segments(gamera, bp[i], "red")
# wait for the user to close the window
turtle.mainloop()
```
